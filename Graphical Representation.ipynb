```python
import numpy as np
from sympy import symbols

def find_integer_points(x_eq, y_eq, x_range=(-7, 7), y_range=(-10, 10)):
    x, y = symbols('x y')
    
    # Use sets to store unique integer points for each equation
    points_x_eq = set()
    points_y_eq = set()
    
    # Generate x-values within the given range
    x_values = np.arange(x_range[0], x_range[1] + 1)
    
    for x_val in x_values:
        # Solve for y using the first equation (where y is a function of x)
        y_val = y_eq.subs(x, x_val)
        
        # Check if the y value is an integer and within the specified range
        if y_val.is_integer and y_range[0] <= y_val <= y_range[1]:
            points_y_eq.add((x_val, int(y_val)))  # Add to the set to ensure uniqueness

        # Solve for x using the second equation (where x is a function of y)
        if len(points_y_eq) > 0:  # Check only if we have valid y values
            y_val_last = points_y_eq.pop()  # Get one valid y value (remove to ensure uniqueness)
            x_val_from_eq = x_eq.subs(y, y_val_last[1])

            # Check if the x value is an integer and within the specified range
            if x_val_from_eq.is_integer and x_range[0] <= x_val_from_eq <= x_range[1]:
                points_x_eq.add((int(x_val_from_eq), int(y_val_last[1])))  # Add to the set

        # Stop once we have 4 unique points for both equations
        if len(points_x_eq) >= 4 and len(points_y_eq) >= 4:
            break

    # Convert sets to lists and limit to 4 points for each equation
    return list(points_x_eq)[:4], list(points_y_eq)[:4]

# Define your equations in isolated form
x, y = symbols('x y')

# Define the equations
y_eq = (3*x)  # First equation: y = 3*x
x_eq = (11 - y)  # Second equation: x = 11 - y

# Get the solution for integer points
integer_points_x_eq, integer_points_y_eq = find_integer_points(x_eq, y_eq)

# Print results for the first equation
print("Integer points for Equation (y = 3*x):")
if integer_points_y_eq:
    for point in integer_points_y_eq:
        print(f"x = {point[0]}, y = {point[1]}")
else:
    print("No valid integer points found for the first equation.")

# Print results for the second equation
print("\nInteger points for Equation (x = 11 - y):")
if integer_points_x_eq:
    for point in integer_points_x_eq:
        print(f"x = {point[0]}, y = {point[1]}")
else:
    print("No valid integer points found for the second equation.")
```
